```python
#load PDF 
from langchain_community.document_loaders import PyPDFLoader
PyPDF = PyPDFLoader('Transformers.pdf')
docs=PyPDF.load()
docs
```

```python
# How to recursively split text by characters [ Text Chunks ]


from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)

#Split all based on chunk size 500 characters and they can be overlap of 50 characters 

final_documents = text_splitter.split_documents(docs) 
```

```python
print(final_documents[0])
print(final_documents[1])
```

```python
#Text splitter > recommended one for generic text. for before example. 
# It is parameterized by a list of characters. It tries to split on them untill chunks are small enough.
#Default list is ["\n\n", "n", " ", ""]. This has the effeect of trying to keep all paragraphs (sentences, and the words)
# How the text is split > By list of characters
# How the chunk size is measured : By number of characters 
```

```python
# Character text split ( character - character Text splitter)

#This is simple method , This splits based on given character sequence, which defaults to "\n\n". chunk
#length is measured by number of characters.

#1, How the text is split; by single character separator.
#2, How the chunk size is measured: by number of characters. 


#Text Loader - Load any .txt file 
from langchain_community.document_loaders import TextLoader
Sloaders = TextLoader('Speech.txt', encoding="utf-8")
docss=Sloaders.load()
docss
```

```python
from langchain_text_splitters import CharacterTextSplitter
text_splitter = CharacterTextSplitter(separator="\n\n", chunk_size=100, chunk_overlap=20)
text_splitter.split_documents(docss)
```

```python
speech=""
with open("Speech.txt",  encoding="utf-8", errors="ignore") as f:
  speech=f.read()
speech

text_splitter=CharacterTextSplitter(chunk_size=100,chunk_overlap=20)
#Good to convert text into documnet type.
text=text_splitter.create_documents([speech])
print(text[0])
print(text[1])
```
