```python
#FAISS > Facebook AI Similarity Search (FAISS) is a library for efficient similarity search and clusterig of dense vectors .
#Its a Vector store DB.

from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter

#loader = TextLoader("Speech.txt")
# Explicitly specify encoding='utf-8'
loader = TextLoader("Speech.txt", encoding='utf-8')

documents=loader.load()

#Text splitter

text_splitter=CharacterTextSplitter(chunk_size=1000, chunk_overlap=30)
docs=text_splitter.split_documents(documents)
```

```python
docs
```

```python
#Ollama embeddings
from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.vectorstores import FAISS
# Initialize embeddings with a specific model
embeddings=OllamaEmbeddings(model="mistral")
#Give all my docs , Then do embeddings for all the docs. And store it in Vector Store DB.
db=FAISS.from_documents(docs,embeddings)
```

```python
## Querying
#Based on speech text
query="How to start snmpd service ?"

# Similarity search, It will similarity search on db for the query.
docs=db.similarity_search(query)
docs[0].page_content


```

```python
# Retriever 
# We can also convert the vectorstore into Retriever class. This allows us to easily use it in other LangChain
# methods, Which largely work with retrievers. (Interface - connected to vectorstore DB)

retriever=db.as_retriever()
docs=retriever.invoke(query)
docs[0].page_content

```

```python
#Similarity search with score 
#There are some FAISS methods, One is similairty_search_with_score, which allow you to return not only the docs' but
#also distance score of the query to them. Lower score is better.

docs_and_score=db.similarity_search_with_score()
#Lower score is good one.
docs_and_score
```
